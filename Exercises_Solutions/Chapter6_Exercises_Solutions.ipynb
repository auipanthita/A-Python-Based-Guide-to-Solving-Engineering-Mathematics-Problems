```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
thai_font = fm.FontProperties(fname='/usr/share/fonts/truetype/tlwg/Laksaman.ttf')
```

```python
#เฉลยแบบฝึกหัดที่ 6.1.1
x = np.linspace(-3, 3, 7)
y = -2 * x + 5

plt.plot(x, y, marker='o')
plt.title('กราฟของฟังก์ชัน $y = -2x + 5$', fontproperties=thai_font)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
#เฉลยแบบฝึกหัดที่ 6.2.1

x = np.linspace(0, 2 * np.pi, 100)

# คำนวณค่า y = |sin(2x)|
y = np.abs(np.sin(2 * x))

# วาดกราฟ
plt.plot(x, y)

# ใส่ชื่อกราฟและแกน
plt.title('กราฟของฟังก์ชัน $y = |sin(2x)|$', fontproperties=thai_font)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# แสดงผล
plt.show()
```

```python
#เฉลยแบบฝึกหัดที่ 6.2.2

# กำหนดช่วงค่า x ตั้งแต่ 0 ถึง 2π โดยแบ่งเป็น 100 จุด
x = np.linspace(0, 2 * np.pi, 100)

# คำนวณค่า y จากฟังก์ชัน cos(x) + sin(2x)
y = np.cos(x) + np.sin(2 * x)

# วาดกราฟ
plt.plot(x, y)
plt.title('กราฟของฟังก์ชัน $cos(x) + sin(2x)$', fontproperties=thai_font)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
#เฉลยแบบฝึกหัดที่ 6.2.3
x = np.linspace(0, 100, 100)

# คำนวณค่า y = sqrt(x)
y = np.sqrt(x)

# วาดกราฟ
plt.plot(x, y)
plt.title(r'กราฟของฟังก์ชัน $y = \sqrt{x}$', fontproperties=thai_font)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
#เฉลยแบบฝึกหัดที่ 6.2.4

x = np.linspace(-3, 3, 100)

# คำนวณค่า y = e^(-x^2)
y = np.exp(-x**2)

# วาดกราฟ
plt.plot(x, y)
plt.title('กราฟของฟังก์ชัน $y = e^{-x^2}$', fontproperties=thai_font)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
#เฉลยแบบฝึกหัดที่ 6.3.1

names = ["สมศักดิ์", "วรินทร", "จิตรลดา", "ธันยา", "อรพรรณ", "กาญจนา"]
scores = [85, 78, 92, 70, 88, 65]

# วาดกราฟแท่ง
plt.bar(names, scores,color='lightpink')

# ตกแต่งกราฟ
plt.title("ผลการสอบวิชาคณิตศาสตร์ของนักศึกษาวิศวกรรม 6 คน", fontproperties=thai_font)
plt.xlabel("ชื่อ", fontproperties=thai_font)
plt.ylabel("คะแนน (เต็ม 100)", fontproperties=thai_font)
#plt.ylim(0, 100)  # ตั้งค่าแกน y ไม่เกิน 100
#plt.grid(axis='y', linestyle='--', alpha=0.7)

# ตั้งค่า font ของ tick labels ให้เป็นภาษาไทย
plt.xticks(fontproperties=thai_font)
plt.yticks(fontproperties=thai_font)

# แสดงผล
plt.show()
```

```python
#เฉลยแบบฝึกหัดที่ 6.3.2
labels = ['Tiktok', 'Instagram', 'Facebook', 'Twitter', 'YouTube']
boy = [3.5, 2.8, 1.5, 1.2, 4.0]   # วัยรุ่นชาย
girl = [4.2, 3.6, 2.1, 0.9, 3.2]   # วัยรุ่นหญิง

# กำหนดตำแหน่งแท่ง
x = np.arange(len(labels))
width = 0.35

# วาดกราฟแท่งคู่
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, boy, width, label='วัยรุ่นชาย', color='skyblue')
rects2 = ax.bar(x + width/2, girl, width, label='วัยรุ่นหญิง', color='lightpink')

# เพิ่มรายละเอียดกราฟ
ax.set_title('กราฟแท่งเวลาการใช้โซเชียลมีเดียต่อวันของวัยรุ่นชายและหญิง (หน่วย: ชั่วโมง)', fontproperties=thai_font)
ax.set_xlabel('แพลตฟอร์ม', fontproperties=thai_font)
ax.set_ylabel('เวลาการใช้โซเชียล (ชั่วโมง)', fontproperties=thai_font)

# ตั้งค่าแกน X ให้ตรงกับ labels
ax.set_xticks(x)
ax.set_xticklabels(labels, fontproperties=thai_font)

# แสดงกราฟ
plt.show()
```

```python
#เฉลยแบบฝึกหัดที่ 6.3.3

# ข้อมูลจากตาราง
labels = ['สุพรรณบุรี', 'ชัยนาท', 'อยุธยา', 'ปทุมธานี', 'นครปฐม']
dryseason = [0.8, 0.7, 0.9, 0.6, 0.8]   # ฤดูแล้ง
rainyseason = [1.2, 1.1, 1.3, 1.0, 1.2]   # ฤดูฝน

# กำหนดตำแหน่งแท่ง
x = np.arange(len(labels))
width = 0.35

# วาดกราฟแท่งคู่
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, dryseason, width, label='ฤดูแล้ง', color='LemonChiffon')
rects2 = ax.bar(x + width/2, rainyseason, width, label='ฤดูฝน', color='LightBlue')

# เพิ่มรายละเอียดกราฟ
ax.set_title('กราฟแท่งผลผลิตของเกษตรกรในฤดูแล้งและฤดูฝน (หน่วย: ตันต่อไร่)', fontproperties=thai_font)
ax.set_xlabel('จังหวัด', fontproperties=thai_font)
ax.set_ylabel('ผลผลิต (ตันต่อไร่)', fontproperties=thai_font)

# ตั้งค่าแกน X ให้ตรงกับ labels
ax.set_xticks(x)
ax.set_xticklabels(labels, fontproperties=thai_font)

# แสดงกราฟ
plt.show()
```
