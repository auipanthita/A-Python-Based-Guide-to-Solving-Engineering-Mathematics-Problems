```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
thai_font = fm.FontProperties(fname='/usr/share/fonts/truetype/tlwg/Laksaman.ttf')
```

```python
#ตัวอย่างที่ 6.1

# กำหนดช่วงของ x
x = np.linspace(-2, 2, 5)

# คำนวณค่า y จากฟังก์ชัน y = 3x + 2
y = 3 * x + 2

# วาดกราฟ
plt.plot(x, y, marker='o')
plt.title('กราฟของฟังก์ชัน $y = 3x + 2$', fontproperties=thai_font)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
#ตัวอย่างที่ 6.2

# กำหนดช่วงของ x
x = np.linspace(-3, 3, 7)

# คำนวณค่า y จากฟังก์ชัน y = -2x + 4
y = -2 * x + 4

# วาดกราฟ
plt.plot(x, y, marker='o')
plt.title('กราฟของฟังก์ชัน $y = -2x + 4$', fontproperties=thai_font)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
#ตัวอย่างที่ 6.3

# กำหนดช่วงของ x
x = np.linspace(-2, 2, 100)

# คำนวณค่า y จากฟังก์ชัน y = e^x
y = np.exp(x)

# วาดกราฟ
plt.plot(x, y)
plt.title('กราฟของฟังก์ชัน $y = e^x$', fontproperties=thai_font)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
#ตัวอย่างที่ 6.4

# กำหนดช่วงของ x จาก 1 ถึง 100
x = np.linspace(1, 100, 100)

# คำนวณค่า y จากฟังก์ชัน y = ln(x)
y = np.log(x)

# วาดกราฟ
plt.plot(x, y)
plt.title(r'กราฟของฟังก์ชัน $y = \ln(x)$', fontproperties=thai_font)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
#ตัวอย่างที่ 6.5
# ข้อมูล
rooms = ['ม.1/1', 'ม.1/2', 'ม.1/3', 'ม.1/4', 'ม.1/5']
math_scores = [75, 82, 78, 85, 80]
eng_scores = [68, 74, 71, 79, 76]

# กำหนดตำแหน่งของแท่ง
x = np.arange(len(rooms))
width = 0.35  # ความกว้างของแท่ง

# วาดกราฟแท่ง
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, math_scores, width, label='คณิตศาสตร์', color='skyblue')
rects2 = ax.bar(x + width/2, eng_scores, width, label='ภาษาอังกฤษ', color='lightgreen')

# เพิ่มรายละเอียดกราฟ
ax.set_title('คะแนนเฉลี่ยวิชาคณิตศาสตร์และภาษาอังกฤษของนักเรียน 5 ห้อง', fontproperties=thai_font)
ax.set_xlabel('ห้อง', fontproperties=thai_font)
ax.set_ylabel('คะแนนเฉลี่ย', fontproperties=thai_font)
ax.set_xticks(x)
ax.set_xticklabels(rooms, fontproperties=thai_font)

# แสดงกราฟ
plt.show()
```

```python
#ตัวอย่างที่ 6.6
# ข้อมูลจากตาราง
labels = ['มีนาคม', 'เมษายน', 'พฤษภาคม', 'มิถุนายน', 'กรกฎาคม']
icecream = [120, 180, 220, 250, 280]   # ยอดขายไอศกรีม
shavedice = [80, 150, 180, 200, 230]   # ยอดขายน้ำแข็งใส

# กำหนดตำแหน่งแท่ง
x = np.arange(len(labels))
width = 0.35

# วาดกราฟแท่งคู่
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, icecream, width, label='ไอศกรีม', color='skyblue')
rects2 = ax.bar(x + width/2, shavedice, width, label='น้ำแข็งใส', color='lightgreen')

# เพิ่มรายละเอียดกราฟ
ax.set_title('กราฟแท่งยอดขายไอศกรีมและน้ำแข็งใสในแต่ละเดือน (หน่วย: บาท)', fontproperties=thai_font)
ax.set_xlabel('เดือน', fontproperties=thai_font)
ax.set_ylabel('ยอดขาย (บาท)', fontproperties=thai_font)

# ตั้งค่าแกน X ให้ตรงกับ labels
ax.set_xticks(x)
ax.set_xticklabels(labels, fontproperties=thai_font)

# แสดงกราฟ
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# โหลดฟอนต์ภาษาไทย
thai_font = fm.FontProperties(fname='/usr/share/fonts/truetype/tlwg/Laksaman.ttf')
```

```python
#ตัวอย่างที่ 6.7

# ข้อมูล
labels = ["วิศวกรรมโยธา", "วิศวกรรมเครื่องกล", "วิศวกรรมไฟฟ้า", 'วิศวกรรมคอมพิวเตอร์']
sizes = [25, 35, 20, 20]  # เปอร์เซ็นต์
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']  # เลือกสีตามต้องการ
explode = (0.05, 0.05, 0.05, 0.05)  # แยกส่วนเล็กน้อยเพื่อเน้น

# วาดกราฟ
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode, textprops={'fontproperties': thai_font})
plt.title("การกระจายพนักงานตามสาขาวิศวกรรม (รวม 200 คน)", fontproperties=thai_font)
plt.show()
```

```python
#ตัวอย่างที่ 6.8

# ข้อมูล
labels = ['ค่าวัสดุ', 'ค่าแรง', 'ค่าออกแบบ', 'ค่าดำเนินการอื่น ๆ']
sizes = [40, 30, 20, 10]

# สร้างกราฟพาย
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, textprops={'fontproperties': thai_font})

# ใส่หัวข้อ
plt.title('การใช้งบประมาณในแต่ละส่วน', fontproperties=thai_font)

# แสดงกราฟ
plt.show()
```
